# 🧪 Ablate AI: Sample Model Training
This notebook demonstrates a standard Random Forest training workflow. We intentionally include some **noisy benchmarks** and **outliers** to test the Ablate AI unlearning performance boost.

```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score

print("Libraries loaded.")
```

```python
# Generate Synthetic Data with intentional outliers (Noisy Samples)
X = np.random.rand(500, 10)
y = (X[:, 0] > 0.5).astype(int)

# Inject 10% Noise (Mislabeled samples)
noise_idx = np.random.choice(500, 50, replace=False)
y[noise_idx] = 1 - y[noise_idx]

print(f"Generated 500 samples with 50 intentional noisy outliers.")
```

```python
# Train Initial Model
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
rf = RandomForestClassifier(n_estimators=100)
rf.fit(X_train, y_train)

y_pred = rf.predict(X_test)
print(f"Baseline Accuracy: {accuracy_score(y_test, y_pred):.4f}")
print(classification_report(y_test, y_pred))
```
